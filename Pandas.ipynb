## 1. Creating a DataFrame from a Dictionary (Default columns Orientation)

```python
import pandas as pd

# Dictionary with lists as values
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}

# Creating DataFrame
df = pd.DataFrame.from_dict(data)

print(df)
```

## 2. Using orient='index' to Create a DataFrame

```python
# Dictionary with nested dictionaries
data = {
    "Alice": {"Age": 25, "City": "New York"},
    "Bob": {"Age": 30, "City": "Los Angeles"},
    "Charlie": {"Age": 35, "City": "Chicago"}
}

# Creating DataFrame with index orientation
df = pd.DataFrame.from_dict(data, orient='index')

print(df)
```

## 3. Creating a DataFrame with orient='columns' (Explicit)

```python
df = pd.DataFrame.from_dict(data, orient='columns')

print(df)
```

## 4. Handling Missing Data

```python
data = {
    "Alice": {"Age": 25, "City": "New York"},
    "Bob": {"Age": 30},  # Missing "City"
    "Charlie": {"City": "Chicago"}  # Missing "Age"
}

df = pd.DataFrame.from_dict(data, orient='index')

print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
import pandas as pd
import numpy as np

data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, 8, 9],
    'C': ['foo', 'bar', 'baz', 'qux', np.nan],
    'D': [np.nan, 3.5, 6.1, np.nan, 2.9],
    'E': [1, 3.5, 6.1, 1, 2.9]
}
df = pd.DataFrame(data)

# Visualizing missing values
plt.figure(figsize=(10, 6))
heatmap = sns.heatmap(df.isnull(), cbar=True, cmap='viridis', yticklabels=True)

# Add labels for the color bar
cbar = heatmap.collections[0].colorbar
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Not Null', 'Null'])

# Add title and labels
plt.title('Heatmap of Missing Values')
plt.xlabel('Columns')
plt.ylabel('Rows')

plt.show()
```

```python

# Creating a DataFrame
data = {'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8], 'C': [2, 4, 6, 8]}
df = pd.DataFrame(data)
print("Original DataFrame:")
display(df)
```

```python

# Using map() with a function on a Series
df['A_mapped_function'] = df['A'].map(lambda x: x * 2)
print("\nmap() with function:")
display(df)
```

```python
# Using map() with a dictionary mapping
dict_map = {1: 'One', 2: 'Two', 3: 'Three', 4: 'Four'}
df['A_mapped_dict'] = df['A'].map(dict_map)
print("\nmap() with dictionary:")
display(df)
```

```python
# Using map() with a Series
df['A_mapped_series'] = df['A'].map(pd.Series({1: 'First', 2: 'Second', 3: 'Third', 4: 'Fourth'}))
print("\nmap() with Series:")
display(df)
```

```python
# Using apply() with a function on a DataFrame
def sum_columns(row):
    return row['A'] + row['B']
```

```python

df['Sum_Apply_Function'] = df.apply(sum_columns, axis=1)
print("\napply() with function:")
display(df)
```
